# Loading images

```python
images = np.zeros((5,128,128,3))
display(images[0].shape)
count = 5
for i in range(count):
    images[i] = plt.imread('images/pic' + str(i) + '.jpg')

plt.imshow(images[0].astype('uint8'))
```

# 2. task, averaging of 5 pictures

```python
averages = np.zeros((128,128,3))
stds     = np.zeros((128,128,3))
for k in range(3):
    for i in range(128):
        for j in range(128):
            averages[i,j,k] = np.average(images[:,i,j,k])
            stds[i,j,k]     = np.std(images[:,i,j,k])
```

```python
plt.imshow(averages.astype('uint8'))
```

```python
display(averages)
display(stds)
```

```python
images_standard = np.zeros((5,128,128,3))
for i in range(count):
    images_standard[i] = (images[i] - averages)/stds
```

# Checking the values

```python
averages = np.zeros((128,128,3))
stds     = np.zeros((128,128,3))
for k in range(3):
    for i in range(128):
        for j in range(128):
            averages[i,j,k] = np.average(images_standard[:,i,j,k])
            stds[i,j,k]     = np.std(images_standard[:,i,j,k])
```

```python
display(averages)
display(stds)
```

# 3. task - audio file

```python
from scipy.io import wavfile

sample_rate, data_raw = wavfile.read('audio/example.wav')

data_left = data_raw[:,0]
data_right = data_raw[:,1]

# mono
data = (data_raw[:,0] + data_raw[:,1]) / 2

fig, axes = plt.subplots(ncols=1, nrows=3, constrained_layout=True)
fig.set_size_inches(12,12)

# Right
axes[0].set_title('Amplitude of the example wav file', fontsize = 18)
axes[0].plot(data_right)
axes[0].set_xlabel('Time')
axes[0].set_ylabel('Amplitude')

axes[1].set_title('Spectrogram of the example wav file', fontsize = 18)
axes[1].specgram(data_right, Fs=sample_rate)
axes[1].set_xlabel('Time')
axes[1].set_ylabel('Frequency')

sp = np.fft.fft(data)
freq = np.fft.fftfreq(data.shape[-1], d=1/sample_rate)
display(freq.shape)
display(sp.shape)

axes[2].set_title('Another type of visualization using Fourier transform', fontsize = 18)
axes[2].plot(freq, sp.real, linewidth=0.2)


plt.show()
```
